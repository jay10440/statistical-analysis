# Use the lm() function to perform linear regression with mpg as response and horsepower as predictor

```r
# Seeing the data
attach(Auto)
```

```r
#Linear Regression:
lm.fit=lm(mpg~horsepower, data=Auto)
lm.fit
summary(lm.fit)
```

```r
#Confidence Interval:
confint(lm.fit)
```

```r
#Predictions:
predict(lm.fit, data.frame(mpg=(c(5,10,15))))
```

```r
#A Scatter Plot With Regression Line
plot(horsepower, mpg)
abline(lm.fit, lwd=3, col="red", pch=20)
```

```r
#Diagnostic Plots
par(mfrow=c(2,2))
plot(lm.fit)
```

```r
# Plotting just the residuals
plot(predict(lm.fit), residuals(lm.fit))
plot(predict(lm.fit), rstudent(lm.fit))
```

Interpretation: Firstly, our prediction line is something like $$y=39.9359-.1578x$$. The slope tells us there is a negative correlation between these two variables, which we've confirmed with our scatterplot. Recall from this data set that y is the mpg and x is horsepower.

Coefficients:
(Intercept)   horsepower  
    39.9359      -0.1578

Next, we see that our $$r^2$$ coefficient is
Multiple R-squared:  0.6059,	Adjusted R-squared:  0.6049

So there's not a super strong correlation, but there is a correlation none-the-less.

We could calculate a prediction based on what we're given: say we have horsepower=98. To compute this, we would simply input it in the equation above:

```r
39.9359-.1578*98
```

We would expect our output to be 24 mpg.

**Confidence Interval** We can see above the associated confidence intervals for the 95%. We are 95% confident that the true means lies between the values shown.
